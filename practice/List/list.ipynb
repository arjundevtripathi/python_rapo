
##  What is a List?
A **List** is a built-in Python data structure that:
- Is **ordered** — items have a defined position (index)
- Is **mutable** — you can add, remove, or change items
- Allows **duplicates** — same value can appear multiple times
- Can hold **mixed data types** — integers, strings, floats, objects, even other lists
- Uses **square brackets** `[]` for creation

```python
my_list = [1, 'hello', 3.14, True, [1, 2]]
```

---
## 📚 All List Methods — Quick Reference

| Method | Description |
|---|---|
| `append(x)` | Add item x to end |
| `extend(iterable)` | Add all items from iterable to end |
| `insert(i, x)` | Insert x at index i |
| `remove(x)` | Remove first occurrence of x |
| `pop(i)` | Remove & return item at index i (default: last) |
| `clear()` | Remove all items |
| `index(x)` | Return index of first occurrence of x |
| `count(x)` | Count occurrences of x |
| `sort()` | Sort list in-place |
| `reverse()` | Reverse list in-place |
| `copy()` | Return shallow copy |

## 📚 Built-in Functions that Work on Lists

| Function | Description |
|---|---|
| `len(lst)` | Number of items |
| `sum(lst)` | Sum of numeric items |
| `min(lst)` | Minimum value |
| `max(lst)` | Maximum value |
| `sorted(lst)` | Returns new sorted list |
| `reversed(lst)` | Returns reversed iterator |
| `enumerate(lst)` | Returns (index, value) pairs |
| `zip(lst1, lst2)` | Pairs items from multiple lists |
| `map(fn, lst)` | Apply function to each item |
| `filter(fn, lst)` | Filter items by function |
| `list(iterable)` | Convert iterable to list |
| `any(lst)` | True if any item is truthy |
| `all(lst)` | True if all items are truthy |

````python
# wap to list comprehension for prime numbers from 1 to 20
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [n for n in range(1, 21) if is_prime(n)]
print("Prime numbers from 1 to 20:", prime_numbers)
````

````python
result_loop = []
for x in range(5):
    result_loop.append(x * 3)


# Equivalent list comprehension
result_comp = [x * 3 for x in range(5)]

print(result_loop)  # [0, 3, 6, 9, 12]
print(result_comp)  # [0, 3, 6, 9, 12]
print(result_loop == result_comp)  # True
````

````python
# wap to swap first and last element of a list
def swap_first_last(lst):
    if len(lst) < 2:
        return lst  # No swap needed for lists with less than 2 elements
    lst[0], lst[-1] = lst[-1], lst[0]
    return lst

print(swap_first_last(["python", "programming"]))  # Output: ["programming", "python"]
print(swap_first_last([1, 2, 3, 4, 5]))  # Output: [5, 2, 3, 4, 1]
print(swap_first_last([10]))  # Output: [10]
print(swap_first_last([]))  # Output: []
````
