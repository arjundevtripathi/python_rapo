## 📌 What is a Tuple?
A **Tuple** is a built-in Python data structure that:
- Is **ordered** — items have a defined position (index)
- Is **immutable** — once created, items cannot be changed, added, or removed
- Allows **duplicates** — same value can appear multiple times
- Can hold **mixed data types** — integers, strings, floats, objects, even other tuples
- Uses **parentheses** `()` for creation (parentheses are optional in most contexts)
- Is **faster** than lists for read-only operations
- Can be used as **dictionary keys** (unlike lists)
- Is **hashable** (if all elements are hashable)

```python
my_tuple = (1, 'hello', 3.14, True, (1, 2))
```

## 🆚 Tuple vs List
| Feature | Tuple | List |
|---|---|---|
| Syntax | `(1, 2, 3)` | `[1, 2, 3]` |
| Mutable | ❌ No | ✅ Yes |
| Speed | ✅ Faster | Slower |
| Dict key | ✅ Yes | ❌ No |
| Memory | ✅ Less | More |
| Methods | 2 | 11 |

## 📚 All Tuple Methods — Quick Reference

Tuples have only **2 built-in methods** (because they are immutable):

| Method | Description | Example |
|---|---|---|
| `count(x)` | Count occurrences of x | `(1,2,1).count(1)` → `2` |
| `index(x)` | Return index of first occurrence of x | `(1,2,3).index(2)` → `1` |

## 📚 Built-in Functions that Work on Tuples

| Function | Description |
|---|---|
| `len(t)` | Number of items |
| `sum(t)` | Sum of numeric items |
| `min(t)` | Minimum value |
| `max(t)` | Maximum value |
| `sorted(t)` | Returns new sorted **list** |
| `reversed(t)` | Returns reversed iterator |
| `enumerate(t)` | Returns (index, value) pairs |
| `zip(t1, t2)` | Pairs items from multiple tuples |
| `map(fn, t)` | Apply function to each item |
| `filter(fn, t)` | Filter items by function |
| `tuple(iterable)` | Convert iterable to tuple |
| `any(t)` | True if any item is truthy |
| `all(t)` | True if all items are truthy |
| `hash(t)` | Return hash value (if all elements hashable) |
| `type(t)` | Returns `<class 'tuple'>` |
